```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = 'class', y = 'fare', data = dataset)
plt.plot()


### A bar represt an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainity around that estimate using error bars.
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = 'class', y = 'fare', data = dataset,
            hue='sex')
plt.plot()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = 'class', y = 'fare', data = dataset,
            hue='sex',
            palette='inferno')
plt.plot()
```

```python
### now if you want to make the horizontal graph instead of this vertical graph 

### then you may need to prove the numeric value for the x 
## like here the value of x is not numeric hence it will not be able to be plotted in the horizontal direction 

## for making the graph into the horizontal direction use 
### orient = 'h'
```

```python
### that vertical line is could be decided by us .. means which thnig you want to calculate

### like we could np.max , np.mean, etc ..
### by default it is np.mean()

plt.figure(figsize=(10,6))
sns.barplot(x = 'class', y = 'fare', data = dataset,
            hue='sex',
            palette='inferno',
            estimator=np.max)
plt.plot()
```

### confidence interval

```python
## we could change the color the width, the length also which is the error indicator .. 

## the maximum value for the ci = 100

plt.figure(figsize=(10,6))
sns.barplot(x = 'class', y = 'fare', data = dataset,
            hue='sex',
            palette='inferno',
            ci=25,
            errcolor='green',
            errwidth=10)
plt.plot()
```

```python
## saturation.. the more value of saturation the more it is colored ,, less value means more to black and white side

plt.figure(figsize=(10,6))
sns.barplot(x = 'class', y = 'fare', data = dataset,
            hue='sex',
            palette='inferno',
            saturation=0.1)
plt.plot()
```
