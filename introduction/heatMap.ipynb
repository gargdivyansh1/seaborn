```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
flight_data = sns.load_dataset('flights')
print(flight_data.columns)
flight_data.head()
```

```python
### there is a method in the pandas for converting the data into the numeric format 
### like here we can convert this data into something like this 

flight_data = flight_data.pivot(index='month', columns='year', values='passengers')
flight_data.head()
```

```python
plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_data)
ax
```

```python
## annotation

plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_data,
                 annot=True,
                 fmt='d')
ax
```

```python
## we can also give the line color and the line width 
## b = represent blue color 
## k = represent black color

plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_data,
                 annot=True,
                 fmt='d',
                 linecolor='k',
                 linewidths=2)
ax
```

```python
### for changing the color of the blocks
plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_data,
                 annot=True,
                 fmt='d',
                 cmap='Blues',
                 linecolor='k',
                 linewidths=2
                 )
ax
```

```python
## if we do not want to have the c bar 

plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_data,
                 annot=True,
                 fmt='d',
                 cmap='Blues',
                 linecolor='k',
                 linewidths=2,
                 cbar=False
                 )
ax
```
