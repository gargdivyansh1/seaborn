```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins)
plt.show()
```

###  What does a pairplot show?


```python
# Scatter plots for each pair of numerical variables.

# Histograms (or KDE plots) on the diagonal to show distributions of each individual variable.

# Optional color/hue grouping based on a categorical variable.
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins, hue='sex', palette="Reds")
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins, hue='species', palette='Blues')
##plt.show()
```

```python
## kind is used when you want to decide how the graph looks like .. and what graph do you want 

plt.figure(figsize=(12,12))
sns.pairplot(penguins,
             hue='species',
             kind='hist', 
             palette='rainbow')
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,
             hue='species',
             diag_kind='hist',
             palette='rainbow')
plt.show()
```

```python
## for only showing the left side data like the bottom triangle below the diagonal

plt.figure(figsize=(12,12))
sns.pairplot(penguins, corner=True)
plt.show()
```

```python
## marker is the concept of the matplotlib only .. it is just used for giving the shapes to the symbols

plt.figure(figsize = (12,12))
sns.pairplot(penguins, hue="species", markers=["o", "s", "D"] , palette = 'inferno')
plt.show()
```
