```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### here we will be using the data sets which are already provided by the seaborn itself using the load_dataset method

dataset = sns.load_dataset('titanic')
```

```python
### making the scatter plot 

plt.figure(figsize=(8,6))
sns.scatterplot(x = 'age' , y = 'fare', data = dataset)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = 'age' , y = 'fare', data = dataset,
                hue='alive')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = 'age' , y = 'fare', data = dataset,
                hue='alive',
                style='class')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = 'age' , y = 'fare', data = dataset,
                hue='alive',
                style='class',
                palette='inferno')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = 'age' , y = 'fare', data = dataset,
                hue='alive',
                style='class',
                palette='gist_rainbow',
                alpha=0.5)
plt.title("Titanic Data Analysis")
plt.show()
```

```python
### here we are drawing the scatter plot and the line plot on the same graph 

plt.figure(figsize=(10,6))
sns.scatterplot(x = 'age' , y = 'fare', data = dataset,
                hue='alive',
                style='class')
sns.lineplot(x = 'age', y = 'fare', data = dataset, 
             hue= 'alive', 
             style='class', 
             color = 'green')
plt.title("Titanic Data Analysis")
plt.show()
```
